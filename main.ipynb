```python
#ejemplo while

temp = 32
print("f°         c°")
while temp < 73:
  print(temp,"   ", int((temp-32)*5/9))
  temp = temp + 1
```

```python
#ejemplo for
print("f°     c°")
for temp in range(21):
  print(temp,"   ",int((temp-32)*5/9))
```

```python
#edades usando for del 1 al 99
for i in range(1,100):
  if i<18:
    print(i,"menor de edad")
  else:
    print(i,"mayor de edad")
```

```python
#edades usando while del 1 al 99
i = 1
while i < 100:
  if i<18:
    print(i,"menor de edad")
  else:
    print(i,"mayor de edad")
  i = i+1
```
